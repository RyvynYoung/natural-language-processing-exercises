```python
# create functions

def get_blog_article_info(url):
    '''
    Takes in the url for article and gets info on the article and creates dictionary of the info
    '''
    # use the make_soup function to create the object from the url
    soup = acquire.make_soup(url)
    # gets these items from the article
    article_title = soup.find('title').text
    article_content = soup.find('div', class_='jupiterx-post-content').text
    article_date = soup.find('time').attrs['datetime']
    # creates a dictionary of the info collected
    art_dict = {'article_title': article_title, 'article_content': article_content, 'article_date': article_date}
    return art_dict

def get_blog_articles(url_list):
    # make an empty list to hold dictionaries
    art_dict_list = []
    #iterate through list of article urls and get info for each article
    for url in url_list:
        art_info = get_blog_article_info(url)
        art_dict_list.append(art_info)
    # use pandas to make a df of the article dictionaries
    article_df = pd.DataFrame(art_dict_list)
    # write the df to local for future use
    article_df.to_json('article_df.json')
    print("article.json file created use pd.read_json('article_df.json') to open the file")
    # for exercises adding return of dictionary list and dataframe
    return art_dict_list, article_df
```

```python
# identify location of info to get: title, content, category, author, date
category = soup.find('li', class_='active-category').text  
title = soup.find('span', itemprop='headline').text
content = soup.find('div', itemprop='articleBody')
author = soup.find('span', class_='author').text
date = soup.find('span', clas='date').text
more_info_url = soup.find('a', class_='source').get('href')
```

```python
def get_inshorts_article_info(url, topic):
    '''
    Takes in the url for article and gets info on the article and creates dictionary of the info
    '''
    # use the make_soup function to create the object from the url
    soup = acquire.make_soup(url)
    # gets these items from the article
    category = topic  
    title = soup.find('span', itemprop='headline').text
    content = soup.find('div', itemprop='articleBody')
    author = soup.find('span', class_='author').text
    date = soup.find('span', clas='date').text
    more_info_url = soup.find('a', class_='source').get('href')
    # creates a dictionary of the info collected
    art_dict = {'category': category, 'article_title': title, 'article_content': content, 
                'article_date': date, 'author': author, 'more_info_url': more_info_url}
    return art_dict


def get_inshorts_articles():
    # list of topics to get articles on 
    topic_list = ['business', 'sports', 'technology', 'entertainment']
    # then a way to format the url to loop through topics
    for topic in topic_list:
        url = f'https://inshorts.com/en/read/{topic}'
        headers = {'User-Agent': 'Codeup Data Science'} # Some websites don't accept the pyhon-requests default user-agent
        response = get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        # make an empty list to hold dictionaries
        art_dict_list = []
        # get list of articles on page
        article_container = soup.find_all('div', class_='new-card z-depth-1')
        # loop through each article and collect info
        for article in article_container:
            art_info = get_inshorts_article_info(url, topic)
            art_dict_list.append(art_info)
    # use pandas to make a df of the article dictionaries
    article_df = pd.DataFrame(art_dict_list)
    # write the df to local for future use
    article_df.to_json('inshorts_df.json')
    print("inshorts.json file created use pd.read_json('inshorts_df.json') to open the file")
    # for exercises adding return of dictionary list and dataframe
    return art_dict_list, article_df
```
